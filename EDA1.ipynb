### Task1

Try to tune the "before" graph to have a "after" plot that is representing the information of
showing the relationship between weight and mpg and highlighting detailed extra information you want to emphasis, and have a good taste of design. Follow the steps below:

- add the regression line to show the linear relationship between Weight and MPG; tune the color, size or style (and etc.) of the line and dots, add label or text to indicate the linear relationship, i.e. MPG=a+b*Weight.
- adjust the position of x label to be on the bottom right, y label to be on the top left (the example in L6 had x label on the bottom left and y label on top left).
- add a 6-12 word descriptive title.
- Pretend the cars with MPG ≥ 40 or Weight ≤ 2000(lb) to be "Our Products", highlight them in the plot (with color or size or both) and add a label or a text of "Our Products" in the position that’s close to them.

```python
df_cars = pd.read_csv("cars.csv")
fig, ax = plt.subplots(figsize=(6,6))
col = np.where((df_cars['MPG']>=40)|(df_cars['WGT']<=2000),'#E15F32','#E2E5EA')
ax.scatter(df_cars['WGT'], df_cars['MPG'], c=col, s=40, alpha=0.6)
#regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(df_cars['WGT'], df_cars['MPG'])
line = intercept+slope*df_cars['WGT']
plt.plot(df_cars['WGT'],line,'black',label='MPG={:.2f}+{:.2f}Weight'.format(slope,intercept))
#title
ax.set_title('Negative correlation between MPG and Weight',fontsize=14)
#label
ax.set_xlabel('Weight')
ax.set_ylabel('MPG')
ax.annotate('Our Products', (2500, 42),color='#E15F32',size=12, weight='bold')
plt.legend(fontsize=10)
ax.xaxis.set_label_coords(0.945, -0.1) 
ax.yaxis.set_label_coords(-0.08, 0.945)
plt.show()
```

```python
df_covid = pd.read_csv('sf_covid.csv')
df_covid['Date'] =pd.to_datetime(df_covid.Date)
df = df_covid.drop(df_covid.columns[[0]],axis = 1)
```

```python
# 'before' plot
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(df['Date'],df['Count'] ) 
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4))
df['month'] = df['Date'].dt.month
df['month_year'] = pd.to_datetime(df['Date']).dt.to_period('M')
df['month_year'] = df['month_year'].astype('str')
df_avg = df.groupby(df['month_year']).mean()
df_min = df.groupby(df['month_year']).min()
df_max = df.groupby(df['month_year']).max()
avg =df_avg['Count']
mi =df_min['Count']
ma =df_max['Count']
plt.xticks(range(12), ('Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','Jan','Feb'))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.plot(avg,linewidth=2.5)
ax.annotate('AVG', (11.2,104),color='#305f91',size=11, weight='bold')
ax.annotate('MAX', (11.2,200),color='#a4a7ab',size=11, weight='bold')
ax.annotate('MIN', (11.2,20),color='#a4a7ab',size=11, weight='bold')
ax.annotate("Average Daily Count of Confirmed Covid Cases (per month)", (-1.2, 600), size=13, color='#262729', annotation_clip=False)
ax.annotate("2020", (3.15, -125), size=9.5, color='#8bb5c7', annotation_clip=False,weight='bold')
#ax.set_xlabel("2020",fontsize=9.5,color='#8bb5c7')
#ax.xaxis.set_label_coords(0.34, -0.13)
ax.set_xlabel("2021",fontsize=9.5,color='black',weight='bold')
ax.xaxis.set_label_coords(0.91, -0.12)
ax.set_ylabel("Average Monthly Count",fontsize=11,color='#262729')
ax.yaxis.set_label_coords(-0.1, 0.65)
plt.fill_between(range(12), mi, ma, facecolor='#a4a7ab',alpha=0.3)
for i in range(0,10):
    ax.get_xticklabels()[i].set_color("#8bb5c7") 
for i in range(10,12):
    ax.get_xticklabels()[i].set_color("black")
```
